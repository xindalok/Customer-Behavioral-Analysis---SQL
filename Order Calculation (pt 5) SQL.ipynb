#### 43. Create a running total of standard_amt_usd in the orders table over order time with no date truncation. The final table should have two columns: one with the amount being added for each new row and a second with the running total.

```python
%%sql 
SELECT 
    standard_amt_usd,
    SUM(standard_amt_usd) OVER(ORDER BY occurred_at) AS running_total
FROM orders_consolidated
LIMIT 100
```

#### 44. Modify the previous query to partition by the year-truncated occurred_at date. The final table should include three columns: the amount being added, the truncated date, and the running total within each year.

```python
%%sql 
SELECT 
    standard_amt_usd,
    EXTRACT(YEAR FROM occurred_at) AS year,
    SUM(standard_amt_usd) OVER(
        PARTITION BY EXTRACT(YEAR FROM occurred_at) 
        ORDER BY occurred_at
    ) AS running_total
FROM orders_consolidated
LIMIT 100
```

#### 45. Rank the total amount of paper ordered (total) for each account (account_id) in the orders table using the RANK() function. The final table should include id, account_id, total, and total_rank.

```python
%%sql 
SELECT 
    RANK() OVER (ORDER BY total_paper_qty DESC) AS ranked,
    *
FROM(
SELECT 
    account_id,
    company,
    SUM(total) AS total_paper_qty
FROM orders_consolidated
GROUP BY account_id,company
ORDER BY total_paper_qty DESC
) AS paper_ranked
LIMIT 100
```

#### 46. Create and use an alias named `account_year_window` to simplify a query containing multiple window functions that partition by `account_id` and order by the truncated year of `occurred_at`.  

```python
%%sql
SELECT 
    account_id,
    DATE_TRUNC('year', occurred_at) AS year,
    SUM(standard_amt_usd) OVER account_year_window AS running_total,
    AVG(standard_amt_usd) OVER account_year_window AS avg_amt
FROM orders
WINDOW account_year_window AS (
    PARTITION BY account_id 
    ORDER BY DATE_TRUNC('year', occurred_at)
)
LIMIT 100
```

#### 47. Compare each order’s total revenue (total_amt_usd) with the next order’s total revenue using the LEAD() function. The final table should include occurred_at, total_amt_usd, lead, and lead_difference.

```python
%%sql 
SELECT 
    occurred_at,
    total_amt_usd,
    LEAD(total_amt_usd, 1) OVER( ORDER BY occurred_at) AS lead,
    total_amt_usd - LEAD(total_amt_usd, 1) OVER( ORDER BY occurred_at) AS lead_difference
FROM orders_consolidated
ORDER BY occurred_at
LIMIT 100
```

#### 48. Use the NTILE(4) function to divide the accounts into four quartiles based on the amount of standard_qty for their orders. The final table should include account_id, occurred_at, standard_qty, and standard_quartile.

```python
%%sql 

SELECT 
    account_id,
    occurred_at,
    standard_qty,
    NTILE(4) OVER(ORDER BY (standard_qty) DESC) AS standard_quartile
FROM orders_consolidated
LIMIT 100
```

#### 49. Use the NTILE(2) function to divide the accounts into two levels based on the amount of gloss_qty for their orders. The final table should include account_id, occurred_at, gloss_qty, and gloss_half.

```python
%%sql
SELECT 
    account_id,
    occurred_at,
    gloss_qty, 
    NTILE(2) OVER(ORDER BY gloss_qty DESC) AS gloss_half
FROM 
orders_consolidated
LIMIT 100 
```

#### 50. Use the NTILE(100) function to divide the orders for each account into 100 percentile levels based on total_amt_usd. The final table should include account_id, occurred_at, total_amt_usd, and total_percentile.

```python
%%sql
SELECT 
    account_id,
    occurred_at,
    total_amt_usd DESC, 
    NTILE(100) OVER(ORDER BY total_amt_usd) AS total_percentile
FROM orders_consolidated
ORDER BY total_amt_usd DESC
LIMIT 100
```

#### 51. Determine how the current order's total revenue ("total" meaning from sales of all types of paper) compares to the next order's total revenue. (There should be four columns: occurred_at, total_amt_usd, lead, and lead_difference.)



```python
%%sql 
SELECT 
    occurred_at,
    total_amt_usd,
    LEAD(total_amt_usd, 1) OVER( ORDER BY occurred_at) AS lead,
    total_amt_usd - LEAD(total_amt_usd, 1) OVER( ORDER BY occurred_at) AS lead_difference
FROM orders_consolidated
ORDER BY occurred_at
LIMIT 100
```
